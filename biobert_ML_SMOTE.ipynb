```python
def remove_tensor(str):
  str=str.replace('tensor','')
  str=str.replace('(','')
  str=str.replace(')','')
  str=str.replace(',','')
  str=str.replace('[','')
  str=str.replace(']','')
  return str
```

```python
result
```

```python

result['sentence embedding'] = result['sentence embedding'] .astype(str)
```

```python
result['sentence embedding']=result['sentence embedding'].apply(remove_tensor)
```

```python
x_val=validation_data.drop('labels' , axis=1) 
y_val=validation_data['labels']
```

```python
x=doc_vectors.drop('labels' , axis=1)
```

```python
y=doc_vectors['labels']
```

**SMOTE**

```python
from imblearn.over_sampling import SMOTE 
```

```python
smote = SMOTE(random_state = 2021)
x, y = smote.fit_resample(x,y)  
```

```python
print(x.shape)
print(y.shape)
print(x_val.shape)
print(y_val.shape)
```

**DECISION TREE**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model= DecisionTreeClassifier(max_leaf_nodes=5)
```

```python
dt_model.fit(x,y)
```

```python
dt_pred=dt_model.predict(x_val) 
```

```python
dt_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print(classification_report(y_val,dt_pred))
```

```python
print(confusion_matrix(y_val,dt_pred))
```

```python
#training accuracy 
dt_pred_train = dt_model.predict(x)
accuracy_score(y,dt_pred_train )  
```

**RANDOM FOREST**

```python
rfc_model=RandomForestClassifier(n_estimators=1000,max_leaf_nodes=50)
```
